```python
import pandas as pd
```

```python
df = pd.read_csv("TTC.csv")
```

```python
df
```

```python
df["distAhead"] = (df.pedX - df.vehX)/10 # in meters
df["distLR"] = (df.pedY - df.vehY)/10 # in meters
```

```python
df
```

```python
import matplotlib.pyplot as plt

x = df[df.TTC==3]["distAhead"]
y = df[df.TTC==3]["distLR"]
plt.xlim(-20, 100)  # Set the lower and upper limits for the x-axis
plt.ylim(-40, 40)  # Set the lower and upper limits for the y-axis
plt.plot([-20, 100], [0, 0], linestyle='dotted', color='gray')
plt.plot([0, 0], [-40, 40], linestyle='dotted', color='gray')
plt.xlabel('Behind << >> Ahead (m)')
plt.ylabel('Right << >> Left (m)')
plt.title('Pedestrian: TTC = 3s')

plt.scatter(x, y)
```

```python
x1 = df[df.TTC==3]["distAhead"].values.tolist()
y1 = df[df.TTC==3]["distLR"].values.tolist()
x2 = df[df.TTC==1]["distAhead"].values.tolist()
y2 = df[df.TTC==1]["distLR"].values.tolist()
plt.xlim(-20, 100)  # Set the lower and upper limits for the x-axis
plt.ylim(-40, 40)  # Set the lower and upper limits for the y-axis
plt.plot([-20, 100], [0, 0], linestyle='dotted', color='gray')
plt.plot([0, 0], [-40, 40], linestyle='dotted', color='gray')
plt.xlabel('Behind << >> Ahead (m)')
plt.ylabel('Right << >> Left (m)')
plt.title('Pedestrian: TTC = 1-3s')

import random

for i in range(len(x1)):
    # Generate a random color
    random_color = "#" + "%06x" % random.randint(0, 0xFFFFFF)

    # Connect the points with lines
    plt.plot([x1[i], x2[i]], [y1[i], y2[i]], linestyle='-', color=random_color)
```
